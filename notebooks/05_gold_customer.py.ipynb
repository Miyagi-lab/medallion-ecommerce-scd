```sql
%python

# Gold Table 1: current customers only
df_current = spark.table(SILVER_SCD_TABLE) \
    .filter(F.col("is_current") == True)
df_current.write.format("delta").mode("overwrite").saveAsTable(GOLD_CURRENT_TABLE)

# Gold Table 2: full history
df_history = spark.table(SILVER_SCD_TABLE)
df_history.write.format("delta").mode("overwrite").saveAsTable(GOLD_HISTORY_TABLE)
```
